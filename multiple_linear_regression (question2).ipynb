```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([
    [2, 15, 40, 0],
    [5, 16, 45, 1],
    [3, 16, 40, 0],
    [10, 18, 50, 5],
    [7, 17, 45, 3],
    [1, 14, 35, 0],
    [8, 16, 45, 4],
    [4, 15, 40, 1],
    [6, 15, 42, 2],
    [12, 19, 55, 8]
])
y = np.array([15, 25, 18, 45, 35, 12, 38, 22, 30, 60])
columns = ['Experience','Education','WeeklyHours','ManagementExperience']
model = LinearRegression()
model.fit(X, y)
```

```python
print("\nQuestion 2.1:\n")
print(f"Intercept (β₀): {model.intercept_:.2f}")
print(f"Coefficients (β₁, β₂, β₃, β₄): {model.coef_}")
```

```python
print("\nQuestion 2.2:\n")

print(f"Prediction equation:")
print(f"Salary = {model.intercept_:.2f} + {model.coef_[0]:.2f} * Experience + {model.coef_[1]:.2f} * Education + {model.coef_[2]:.2f} * WeeklyHours + {model.coef_[3]:.2f} * ManagementExperience")
```

```python
print("\nQuestion 2.3:\n")

from sklearn.metrics import r2_score

y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print(f"R²: {r2:.4f}")

n = len(y)
p = X.shape[1]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print(f"Adjusted R²: {adj_r2:.4f}")
```

```python
print("\nQuestion 2.4:\n")

def new_worker_salary_calculator(name, new_worker):

    yoex, yoed, wwh, yom = new_worker[name][0]

    new_worker_salary = model.predict(new_worker[name])[0]
    return yoex,yoed,wwh,yom,new_worker_salary
```

```python
name = "Tom"
new_worker = {name:[[6, 16, 43, 2]]}
yoex,yoed,wwh,yom,new_worker_salary = new_worker_salary_calculator(name, new_worker)

print(f"Hiring {name} as a new worker with the following details:\
\n - Years of Experience: {yoex}\n - Years of Education: {yoed}\
\n - Weekly Working Hours : {wwh}\n - Years of Management Experience: {yom}")
print(f"According to his qualifications and experience, the salary for our new worker, {name}, should be {new_worker_salary:.2f}k")
```

```python
print("\nQuestion 2.5:\n")

print("According to out findings  in Question 2.1, tht most influential features affecting a worker's salary are ranked as follows:")
print(f" 1. Management of Experience - β = {model.coef_[3]:.2f}")
print(f" 2. Years of Experience - β = {model.coef_[0]:.2f}")
print(f" 3. Weekly Working Hours - β = {model.coef_[2]:.2f}")
print(f" 4. Years of Education - β = {model.coef_[1]:.2f}")
print("\nNote: The β coefficient represents the weight of each feature in the final salary calculation.\
\nThe higher the value of β, the more influence that feature has.")
```
